```python
df=data[['1.2','2612']]
```

```python
import sklearn.cluster as cluster
```

```python
K = range(1,12)
wss=[]
for k in K:
    kmeans=cluster.KMeans(n_clusters=k,init="k-means++")
    kmeans=kmeans.fit(df)
    wss_iter=kmeans.inertia_
    wss.append(wss_iter)
```

```python
mycenters= pd.DataFrame({'Cluster':K,'WSS':wss})
mycenters
```

```python
#sns.scatterplot(x='Cluster',y="WSS",data=mycenters,marker="+",color="Black")

wss=[] # within cluster sum of squares

for k in range(1, 12):
    kmeansForLoop = KMeans(n_clusters = k)
    kmeansForLoop.fit(df)
    wss.append(kmeansForLoop.inertia_)

plt.figure(figsize = (10, 10))
plt.plot(range(1, 12), wss)
plt.xlabel("K value")
plt.ylabel("WCSS")
plt.show()
```

```python
#sns.scatterplot(x='1.2',y='2612',hue='Cluster',data=df)
import sklearn.metrics as metrics
from sklearn.metrics import silhouette_score
silhouette_score(df,kmeans.labels_)
```

```python


rcParams['figure.figsize']=16,5
_ = plt.plot(range(2,10),silhouette_score,"bo-",color = 'blue',linewidth=3,markersize=8,label='silhouette curve')
_ = plt.xlabel("$k$",fontsize=14,family='Arial')
_ = plt.ylabel("Silhouette curve",fontsize=14,family='Arial')
_ = plt.grid(which='major',color='#cccccc',linestyle='--')
_ = plt.title("Silhouette curve for optimal clusters", family='Arial', fontsize=14)

k = np.argmax(silhouette_score) + 2

_ = plt.axvline(x=k,linestyle='--',c='green',linewidth=2,label='Optimal number of clusters ({})'.format(k))
_ = plt.scatter(k, silhouette_score[k-2],c='red',s=400)
_ = plt.legend(shadow=True)
_ = plt.show()
```
